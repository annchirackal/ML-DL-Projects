Step2:Importing Data and preparing Data

```python
def datapreprocessing(filepath): ## function for reading file and normalising and spliting input and output
    data=pd.read_csv(filepath)
    x=data.iloc[:,1:] # splitting input and output
    y=data['label']
    x=x/255
    return x,y
```

```python
def datasplit(x,y):## spliting training and validation data
    x_train, x_validation, y_train, y_validation = train_test_split(x, y,test_size=.2,random_state=1)#spliting training data int
    return(x_train, x_validation, y_train, y_validation)
```

```python
filepath = "C:/Users/Ann/train.csv"
x,y=datapreprocessing(filepath)
x_train, x_validation, y_train, y_validation=datasplit(x,y)
x_train, x_test, y_train, y_test=datasplit(x_train,y_train)
```

```python
x_train.head()
```

```python
model = keras.models.Sequential()
input_1 = keras.layers.Input(shape=x_train.shape[1])
hidden_1 = keras.layers.Dense(600, activation="relu")(input_1)
hidden_2= keras.layers.Dense(600, activation="relu")(hidden_1)
hidden_3= keras.layers.Dense(600, activation="relu")(hidden_2)
hidden_4=keras.layers.Dense(600, activation="relu")(hidden_3)
hidden_5=keras.layers.Dense(600, activation="relu")(hidden_4)
hidden_6 = keras.layers.Dense(10, activation="softmax")(hidden_4)
model = keras.models.Model(inputs=[input_1], outputs=[hidden_6])
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train, epochs=10,
                    validation_data=(x_validation, y_validation))
```
